# 3.DATA PREPROCESSING

```python
#Dropping columns, each one has no direct association with stroke
data=data.drop(["id"],axis=1)
data
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
lnc=LabelEncoder()
```

```python
data["gender"]=lnc.fit_transform(data["gender"])
data["ever_married"]=lnc.fit_transform(data["ever_married"])
data["work_type"]=lnc.fit_transform(data["work_type"])
data["Residence_type"]=lnc.fit_transform(data["Residence_type"])
data["smoking_status"]=lnc.fit_transform(data["smoking_status"])
```

```python
data
```

# 4.MODELLING AND EVALUATION


```python
x=data.drop("stroke",axis=1)
y=data["stroke"]
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
x_train,y_train = sm.fit_resample(x_train,y_train)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=y_train)
plt.title("Target Distribution AFTER SMOTE")
plt.xlabel("stroke")
plt.ylabel("Count")
plt.show()
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
log_model=LogisticRegression()
log_model.fit(x_train,y_train)
print("train_data_score:",log_model.score(x_train,y_train))
print("test_data_score:",log_model.score(x_test,y_test))   
```

```python
model_rand=RandomForestClassifier(n_estimators=23,random_state=90)
model_rand.fit(x_train,y_train)
print("train_data_score:",model_rand.score(x_train,y_train))
print("test_data_score:",model_rand.score(x_test,y_test)) 
```

```python
knnmodel=KNeighborsClassifier(n_neighbors=10)
knnmodel.fit(x_train,y_train)
print("train_data_score:",knnmodel.score(x_train,y_train))
print("test_data_score:",knnmodel.score(x_test,y_test)) 
```

```python
model_dec=DecisionTreeClassifier()
model_dec.fit(x_train,y_train)
print("train_data_score:",model_dec.score(x_train,y_train))
print("test_data_score:",model_dec.score(x_test,y_test)) 
```

```python
admodel=AdaBoostClassifier(n_estimators=500,estimator=DecisionTreeClassifier(),random_state=2)
admodel.fit(x_train,y_train)
print("train_data_score:",admodel.score(x_train,y_train))
print("test_data_score:",admodel.score(x_test,y_test))
```

```python
model_Gua=GaussianNB()
model_Gua.fit(x_train,y_train)
print("train_data_score:",model_Gua.score(x_train,y_train))
print("test_data_score:",model_Gua.score(x_test,y_test)) 
```

# SAVING MODEL

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', RandomForestClassifier())
])

pipeline.fit(x_train, y_train)
```

```python
# Save the pipeline
with open('stroke_prediction_model.pkl', 'wb') as file:
    pickle.dump(pipeline, file)
```
